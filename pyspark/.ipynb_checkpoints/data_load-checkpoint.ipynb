# Load data from postgre DB

```python
import os

JDBC_PATH = "/home/jovyan/work/libs/postgresql-42.7.5.jar"
JDBC_URL = "jdbc:postgresql://postgres_db:5432/postgres"

POSTGRES_USER = os.getenv("POSTGRES_USER")
POSTGRES_PASSWORD = os.getenv("POSTGRES_PASSWORD")

print(POSTGRES_USER)
```

```python
from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("Postgres-PySpark") \
    .config("spark.driver.extraClassPath", JDBC_PATH) \
    .getOrCreate()

print("Success to create spark session ...")
```

```python
# # Load Data from PostgreSQL
df = spark.read.format("jdbc") \
            .option("url", JDBC_URL) \
            .option("driver", "org.postgresql.Driver") \
            .option("dbtable", "telematics") \
            .option("user", POSTGRES_USER) \
            .option("password", POSTGRES_PASSWORD) \
            .load()

# Show Data
df.show()
```
